```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

'''
Student    Marks     Grade
1          95        85
2          85        95
3          80        70
4          70        65
5          60        70
'''

# Convert table into 2D Matrix
A = [[1,95],[1,85],[1,80],[1,70],[1,60]];
y = [85, 95, 70, 65, 70]

# Convert to numpy array
A = np.array(A)
y = np.array(y)

# Calculate A.T*A
aTa = np.dot(A.T, A)

# Calculate (A.T * A)^-1
aTa_inv = np.linalg.inv(aTa)

# Calculate A.T * y
aT_y = np.dot(A.T, y)

# Calculate theta
theta = np.dot(aTa_inv, aT_y)
print("Theta values:", theta)

# Predict additional values
x_test = [65, 75, 77, 83, 87]
x_test = np.array(x_test)

# Make predictions using h(theta) = theta_0 + theta_1 * x_1
theta_0 = theta[0]
theta_1 = theta[1]
y_pred = theta_0 + theta_1 * x_test
print("Predicted values:", y_pred)

# Plot data and regression line
plt.xlabel = ('Test Marks')
plt.ylabel = ('Grades')
plt.xlim([55, 100])
plt.ylim([55, 100])
plt.scatter(A[:, 1], y)
plt.scatter(A[:, 1], y, color='blue', label='Data Points')
plt.scatter(x_test, y_pred, color='green', label='Predicted Values')
plt.plot(A[:, 1], np.dot(A, theta), color='red', label='Regression Line')
plt.legend()
plt.title('Test Marks vs. Student Grades')
plt.show()
```
